```python
road = pd.get_dummies(dataset['Road_condition'], drop_first=True)
address = pd.get_dummies(dataset['Adress_type'], drop_first=True)
collision = pd.get_dummies(dataset['Collision_type'], drop_first=True)
ttdata = dataset[['SEVERITYCODE', 'Person_count']]
ttdata = pd.concat([ttdata, road, address, collision], axis = 1)
ttdata.dtypes
```

```python
X = np.asarray(ttdata[['Person_count', 'Ice','Oil','Other','Sand/Mud/Dirt','Snow/Slush','Standing Water','Unknown','Wet','Block','Intersection','Cycles', 'Head On','Left Turn', 
                         'Other','Parked Car','Pedestrian','Rear Ended','Right Turn','Sideswipe' ]])
X[0:5]

y = np.asarray(ttdata['SEVERITYCODE'])
y [0:5]
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
LR
```

```python
yhat = LR.predict(X_test)
yhat
yhat_prob = LR.predict_proba(X_test)
yhat_prob
```

```python
from sklearn.metrics import jaccard_similarity_score
jaccard_similarity_score(y_test, yhat)
```

```python
from sklearn.metrics import classification_report
print (classification_report(y_test, yhat))
```

```python
from sklearn.neighbors import KNeighborsClassifier

k = 4 
neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)
neigh

yhat2 = neigh.predict(X_test)
yhat2[0:5]

from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_train, neigh.predict(X_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat2))

Ks = 10
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))
ConfustionMx = [];
for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat2=neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat2)

    
    std_acc[n-1]=np.std(yhat2==y_test)/np.sqrt(yhat2.shape[0])

mean_acc
```

```python
plt.plot(range(1,Ks),mean_acc,'g')
plt.fill_between(range(1,Ks),mean_acc - 1 * std_acc,mean_acc + 1 * std_acc, alpha=0.10)
plt.legend(('Accuracy ', '+/- 3xstd'))
plt.ylabel('Accuracy ')
plt.xlabel('Number of Nabors (K)')
plt.tight_layout()
plt.show()

print( "The best accuracy was with", mean_acc.max(), "with k=", mean_acc.argmax()+1) 
```
